```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = yf.download("MUSA", start="2020-01-01", end="2020-06-01", auto_adjust=False)
b = yf.download("^SP500TR", start="2020-01-01", end="2020-06-01", auto_adjust=False)
c = yf.download("CASY", start="2020-01-01", end="2020-06-01", auto_adjust=False)
d =yf.download("8278.T", start="2020-01-01", end="2020-06-01", auto_adjust=False)
e = yf.download("ATD.TO", start="2020-01-01", end="2020-06-01", auto_adjust=False)
f = yf.download("SVNDY", start="2020-01-01", end="2020-06-01", auto_adjust=False)
g = yf.download("5903.TWO", start="2020-01-01", end="2020-06-01", auto_adjust=False)
# h = yf.download("TSLA", start="2020-01-01", end="2020-06-01", auto_adjust=False)
i = yf.download("LMT", start="2020-01-01", end="2020-06-01", auto_adjust=False)

# Take all days of adjusted closing prices for both stocks
closing_a = a["Adj Close"][-10000:]
closing_b = b["Adj Close"][-10000:]
closing_c = c["Adj Close"][-10000:]
closing_d = d["Adj Close"][-10000:]
closing_e = e["Adj Close"][-10000:]
closing_f = f["Adj Close"][-10000:]
closing_g = g["Adj Close"][-10000:]
# closing_h = h["Adj Close"][-10000:]
closing_i = i["Adj Close"][-10000:]

# Normalize
norm_a = closing_a / closing_a.iloc[0]
norm_b = closing_b / closing_b.iloc[0]
norm_c = closing_c / closing_c.iloc[0]
norm_d = closing_d / closing_d.iloc[0]
norm_e = closing_e / closing_e.iloc[0]
norm_f = closing_f / closing_f.iloc[0]
norm_g = closing_g / closing_g.iloc[0]
# norm_h = closing_h / closing_h.iloc[0]
norm_i = closing_i / closing_i.iloc[0]
```

```python
window = 5 # Rolling window size
plt.plot(norm_a.rolling(window).mean(), label="Murphy USA", color="blue", linewidth=2)
plt.plot(norm_b.rolling(window).mean(), label="S&P500 (SPY)", color="red", linewidth=2)
plt.plot(norm_c.rolling(window).mean(),label="Caseys General Stores", color = "yellow", linewidth=2)
plt.plot(norm_d.rolling(window).mean(),label="Fuji Co", color="purple", linewidth=2)
plt.plot(norm_e.rolling(window).mean(),label="Alimentation Couche Tard", color = "black", linewidth=2)
plt.plot(norm_f.rolling(window).mean(),label="Seven Eleven", color="green", linewidth=2)
#plt.plot(norm_g.rolling(window).mean(),label="Family Mart", color="brown", linewidth=2)
# plt.plot(norm_h.rolling(window).mean(),label="Tesla", color="orange", linewidth=2)

plt.title("RDA")
plt.xlabel("Date")
plt.ylabel("Adjusted Close Price / Initial Close Price")
plt.legend()
plt.style.use("ggplot")

plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Download stock data
sp500 = yf.download("^SP500TR", start="2020-01-01", end="2020-06-01", auto_adjust=False)["Adj Close"]
tsla = yf.download("TSLA", start="2020-01-01", end="2020-06-01", auto_adjust=False)["Adj Close"]
svnelvn = yf.download("SVNDY", start="2020-01-01", end="2020-06-01", auto_adjust=False)["Adj Close"]
print(tsla.head())

# Normalize (Start = 1)
sp500_norm = sp500 / sp500.iloc[0]
tsla_norm = tsla / tsla.iloc[0]
svnelvn_norm = svnelvn / svnelvn.iloc[0]

# Plot
plt.figure(figsize=(10, 5))
plt.plot(sp500_norm, label="S&P 500", color="black", linestyle="--")
plt.plot(tsla_norm, label="Tesla", color="red")
plt.plot(svnelvn_norm, label="Seven Eleven", color="green")
plt.xlabel("Date")
plt.ylabel("Normalized Price")
plt.title("Tesla vs. S&P 500 During COVID Recession")
plt.legend()
plt.show()
```
